```python
import random
import numpy as np
from joblib import Parallel, delayed
from tqdm.notebook import trange
```

```python
# Let 11: J, 12: Q, 13: K, 14: A
deck = list(range(2, 15))
deck *= 4
assert len(deck) == 52
```

```python
def fiddler_slap():
    random.shuffle(deck)
    them = deck[0::2] # They get the even-indexed cards
    # me = deck[1::2] # I get the odd-indexed cards
    counts = np.bincount(them, None, 15)
    face_cards = counts[-4:]
    tester = all(face_cards >= 3)

    return tester
```

```python
TRIALS = 10**7
mc = Parallel(n_jobs=-2)(delayed(fiddler_slap)() 
                         for _ in trange(TRIALS, unit_scale=True))
```

```python
m = sum(mc) / TRIALS
print(f'{m:.2%}')
```
