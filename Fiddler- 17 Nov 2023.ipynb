```python
import numpy as np
from joblib import Parallel, delayed
```

```python
expected = (1, 3, 3, 1) # Expected instances of 0, 1, 2, & 3 Heads respectively
```

```python
def fidler_coin_toss(coins=3, trials=8):
    flips = np.random.binomial(coins, 0.5, trials)
    tab = tuple(np.bincount(flips, None, 4))
    tester = tab == expected
    
    return tester
```

```python
runs = 10**7
dgen = (delayed(fidler_coin_toss)(3, 8) for _ in range(runs))
mc = Parallel(n_jobs=-1, verbose=2)(dgen)
```

```python
m = sum(mc) / runs
print(f'{m:.3%}')
```
