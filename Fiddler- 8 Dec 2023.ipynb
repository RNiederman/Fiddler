```python
import random
import numpy as np
from joblib import Parallel, delayed
from tqdm.notebook import trange
```

```python
def fiddler_connections():
    groups = ['A', 'B', 'C', 'D'] * 4
    draw = random.sample(groups, k=4)
    a, b = np.unique(draw, return_counts=True)
    mx = max(b)

    return mx
```

```python
RUNS = 10**7
```

```python
mc = Parallel(n_jobs=-2)(delayed(fiddler_connections)() 
                         for _ in trange(RUNS, unit_scale=True))
```

```python
bins = np.bincount(mc)
threes = bins[3]
m = threes / RUNS
print(f'{m:.2%}')
```
